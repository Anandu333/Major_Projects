```python
df = pd.DataFrame({'user_message': messages, 'message_date': dates})

# Convert message_date to datetime
try:
    df['message_date'] = pd.to_datetime(df['message_date'], format="%m/%d/%y, %I:%M%p - ")
except ValueError:
    try:
        df['message_date'] = pd.to_datetime(df['message_date'], format= "%d/%m/%Y, %I:%M %p - ")
    except ValueError:
        try:
            df['message_date'] = pd.to_datetime(df['message_date'], format="%d/%m/%y, %I:%M%p - ")
        except ValueError:
            try:
                df['message_date'] = pd.to_datetime(df['message_date'], format="%d/%m/%Y, %I:%M%p - ")
            except ValueError:
            
                print("Unable to parse dates. Invalid date formats.")
        
        

# Rename 'message_date' column to 'date'
df.rename(columns={'message_date': 'date'}, inplace=True)

# Display DataFrame
df.head()
```

```python
df.shape
```

# Data Preprocessing

```python
# Seperate users and messages
users=[]
messages= []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]: #user name
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user']= users
df['message']= messages
df.drop(columns=['user_message'],inplace=True)

df.head()
```

```python
df[df['user']== 'group_notification']
```

```python
df.drop(df[df['user'] == 'group_notification'].index, inplace=True)
df.head()
```

6.To fetch most used words

```python
# remove media ommited messages
# remove stopwords


temp = df[df['message'] != '<Media omitted>\n']
```

```python
f= open('stop_hinglish.txt', 'r')
stop_words= f.read()
```

```python
words=[]

for message in temp['message']:
    for word in message.lower().split():
        if word not in stop_words:
            words.append(word)
    
```

```python
from collections import Counter
new_df= pd.DataFrame(Counter(words).most_common(20))
new_df.rename(columns={0: 'Word',1:'count'}, inplace= True)
```

12. To make a heatmap for checking most active hours.

```python
#Period converts the hour into 10-11 fromat
period= []
for hour in df[['day_name', 'hour']]['hour']:
    if hour ==23:
        period.append(str(hour)+"-"+ str('00'))
    elif hour==0:
        period.append(str('00')+ "-"+ str(hour+1))
    else:
        period.append(str(hour)+"-"+ str(hour+1))
    
```

```python
#To get a column of period in df.
df['period']= period
```

```python
df
```

```python
import seaborn as sns
plt.figure(figsize=(20,6))
sns.heatmap(df.pivot_table(index='day_name',columns='period',values='message',aggfunc='count').fillna(0))
plt.yticks(rotation='horizontal')
plt.show()
```
